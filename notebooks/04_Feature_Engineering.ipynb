# Customer Churn Prediction

# 4. Feature Engineering

## Business Objective

Feature engineering is the process of transforming raw data into a format suitable for machine learning algorithms.

The objective of this phase is to:

- Prepare the dataset for model training.
- Convert categorical variables into numerical format.
- Remove irrelevant features.
- Scale numerical variables where necessary.
- Save the processed dataset for machine learning.

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../data/processed/cleaned_telco_customer_churn.csv")

df.head()
```

```python
df.info()
```

## 4.1 Remove Identifier Column

### Why remove `customerID`?

The `customerID` column is a unique identifier and does not contain predictive information about customer churn.

Removing identifier columns helps the model focus on meaningful customer attributes.

```python
# Remove customerID column

df = df.drop(columns=["customerID"])

df.head()
```

```python
df.columns
```

## 4.2 Separate Features and Target

Machine learning models learn patterns from input features (`X`) to predict the target variable (`y`).

- **Features (`X`)** represent customer attributes.
- **Target (`y`)** represents whether the customer churned.

```python
# Separate features and target

X = df.drop(columns=["Churn"])

y = df["Churn"]
```

```python
print("Features Shape :", X.shape)
print("Target Shape :", y.shape)
```

## 4.3 Encode the Target Variable

### Why encode the target variable?

The target variable (`Churn`) contains categorical values (`Yes` and `No`).

Machine learning algorithms require numerical input. Therefore, the target variable is converted into binary values:

- **No → 0**
- **Yes → 1**

This binary representation is suitable for classification models.

```python
# Encode target variable

y = y.map({
    "No": 0,
    "Yes": 1
})

y.head()
```

```python
y.value_counts()
```

## 4.4 Encode Categorical Features

### Why encode categorical features?

Machine learning algorithms require numerical input.

Most categorical features in this dataset are nominal, meaning they do not have a meaningful numerical order.

Therefore, One-Hot Encoding is used to convert categorical variables into binary indicator columns while avoiding artificial numerical relationships.

```python
# Identify categorical columns

categorical_columns = X.select_dtypes(include=["object"]).columns

categorical_columns
```

```python
# Apply One-Hot Encoding

X = pd.get_dummies(
    X,
    columns=categorical_columns,
    drop_first=True,
    dtype=int
)

X.head()
```

```python
print("Shape after encoding:", X.shape)
```

## 4.5 Feature Scaling

### Why perform feature scaling?

Feature scaling transforms numerical features so that they are on a similar scale.

Algorithms such as Logistic Regression, K-Nearest Neighbors, Support Vector Machines, and Neural Networks are sensitive to feature magnitudes and generally perform better when features are scaled.

Tree-based models such as Decision Trees, Random Forests, and XGBoost do not require feature scaling.

```python
# Numerical columns

numerical_columns = [
    "SeniorCitizen",
    "tenure",
    "MonthlyCharges",
    "TotalCharges"
]

numerical_columns
```

```python
# Create a copy before scaling

X_scaled = X.copy()
```

```python
scaler = StandardScaler()

X_scaled[numerical_columns] = scaler.fit_transform(
    X_scaled[numerical_columns]
)

X_scaled.head()
```

```python
comparison = pd.DataFrame({
    "Original": X[numerical_columns].iloc[0],
    "Scaled": X_scaled[numerical_columns].iloc[0]
})

comparison
```

```python
# Save processed datasets

X.to_csv("../data/processed/X_unscaled.csv", index=False)
X_scaled.to_csv("../data/processed/X_scaled.csv", index=False)
y.to_csv("../data/processed/y.csv", index=False)
```

# 4.6 Save Preprocessing Objects

## Why Save the Preprocessing Objects?

To ensure consistent predictions during deployment, the preprocessing objects used during model training must be preserved.

The following objects are saved:

- **StandardScaler** – Stores the scaling parameters (mean and standard deviation) learned from the training data. This ensures that new customer data is transformed using the exact same scaling process before making predictions.

- **Feature Column Names** – Stores the order and names of the features used during model training. This guarantees that future input data matches the structure expected by the trained machine learning model.

Saving these preprocessing objects enables the model to be deployed in applications such as Streamlit without introducing inconsistencies between training and prediction.

```python
import joblib

# Save the fitted scaler
joblib.dump(
    scaler,
    "../models/scaler.pkl"
)

print("Scaler saved successfully!")
```

```python
# Save feature column names

joblib.dump(
    X_scaled.columns.tolist(),
    "../models/feature_columns.pkl"
)

print("Feature columns saved successfully!")
```
