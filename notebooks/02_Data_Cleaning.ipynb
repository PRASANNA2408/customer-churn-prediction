## 2. Data Cleaning

### Business Objective

The objective of this phase is to identify and resolve data quality issues before analysis and machine learning. Clean and reliable data improves the accuracy of insights and predictive models.

The following tasks will be performed:

- Check duplicate records
- Identify missing values
- Detect hidden missing values
- Correct incorrect data types
- Prepare the dataset for exploratory data analysis (EDA)

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("E:/data_science/projects/customer_churn_prediction/data/raw/telco_customer_churn.csv")
```

## 2.1 Check for Duplicate Records

Duplicate records can lead to incorrect analysis and biased machine learning models. Before cleaning the dataset, we verify whether duplicate customer records exist.

```python
# Check duplicate records
duplicate_count = df.duplicated().sum()

print(f"Number of duplicate records: {duplicate_count}")
```

## 2.2 Check Missing Values

```python
# Check missing values
df.isnull().sum()
```

## 2.3 Investigate Hidden Missing Values

```python
# Check for blank values in TotalCharges
(df["TotalCharges"] == " ").sum()
```

```python
# Display rows where TotalCharges is blank
df[df["TotalCharges"] == " "]
```

```python
df[df["TotalCharges"] == " "][["customerID", "tenure", "MonthlyCharges", "TotalCharges", "Churn"]]
```

```python
import numpy as np

df["TotalCharges"] = df["TotalCharges"].replace(" ", np.nan)
```

```python
df.isnull().sum()
```

### Observation

Although `df.info()` and `df.isnull().sum()` initially indicated no missing values, further investigation revealed that the `TotalCharges` column contained 11 blank string values (`" "`).

These blank strings were converted to `NaN` so they could be properly identified and handled during the data cleaning process.

```python
df[df["TotalCharges"].isnull()]
```

```python
df[df["TotalCharges"].isnull()][[
    "customerID",
    "tenure",
    "MonthlyCharges",
    "Contract",
    "Churn"
]]
```

### 2.4 Handling Missing Values

The `TotalCharges` column contained 11 hidden missing values represented as blank strings. These were converted to `NaN`.

Since these records account for only 11 out of 7,043 customers (approximately 0.16% of the dataset), removing them has a negligible impact on the analysis while preserving data quality.

```python
# Remove rows with missing TotalCharges
df = df.dropna(subset=["TotalCharges"])

# Check dataset shape
df.shape
```

### 2.5 Convert Data Types

The `TotalCharges` column was originally stored as a string because of hidden blank values. After removing those records, the column is converted to a numeric data type for further analysis.

```python
# Convert TotalCharges to numeric
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"])
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.to_csv(
    "E:/data_science/projects/customer_churn_prediction/data/processed/cleaned_telco_customer_churn.csv",
    index=False
)
```
